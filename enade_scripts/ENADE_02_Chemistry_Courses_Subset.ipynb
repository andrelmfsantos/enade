## Ação afirmativa relacionada ao financiamento governamental

```python
# Ação Afirmativa
df_gov = df_quimica[['sit_financeira','acao_afirmativa']]#.drop_duplicates()

df_gov = df_gov.groupby(['sit_financeira','acao_afirmativa']).size().reset_index()
df_gov.columns = ['sit_financeira','acao_afirmativa','size']
df_gov['Freq'] = df_gov['size']/df_gov['size'].sum()*100
df_gov.round(2).sort_values(by = ['Freq'], ascending = False)#.head()
```

```python
# @title Barplot :: Distribuição de Vagas Preenchidas pelas Reservas para Estudantes por Critérios da Ação Afirmativa
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate relative frequency
df_gov1 = df_gov.groupby(['acao_afirmativa'])[['size']].agg(sum).reset_index()
total_students = df_gov1['size'].sum()
df_gov1['relative_frequency'] = df_gov1['size'] / total_students

# Sort the DataFrame by 'num_students' for correct bar order
df_sorted = df_gov1.sort_values('size', ascending=False)
print(df_sorted)

# Create a horizontal bar plot with catplot
g = sns.catplot(x='size', y='acao_afirmativa', data=df_sorted, kind='bar', orient='h', height=5, aspect=2)
g.set_axis_labels("Number of Students", "Class")

# Add labels for relative frequency in the specified order
for i, (index, row) in enumerate(df_sorted.iterrows()):
  g.ax.text(row['size'], i, f"{row['relative_frequency']:.2%}", color='black', ha="left", va="center")


# Customizing the plot
plt.xlabel('Número de alunos')
plt.ylabel('Critérios da Ação Afirmativa')
#Number of Students in Each Class with Relative Frequency Labels
plt.title('Distribuição de Estudantes por Vagas Reservadas por Tipo de Ação Afirmativa')
plt.tight_layout()

# Show the plot
plt.show()
```
